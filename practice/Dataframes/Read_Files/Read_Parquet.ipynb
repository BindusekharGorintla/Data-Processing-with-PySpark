```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Read parquet").getOrCreate()
```

```python
df_parquet = spark.read.parquet("/Volumes/workspace/practice/my_volume/employees_parquet")
df_parquet.show()
```

```python
df_parquet.printSchema()
```

```python
df_parquet.write.mode("overwrite").saveAsTable("workspace.practice.employees_parquet")
```
