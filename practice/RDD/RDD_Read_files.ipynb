```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("RDD read files").getOrCreate()
```

```python
rdd = spark.sparkContext.textFile(...) will only run on standard Databricks clusters (single-user or multi-user clusters) where you have full Spark driver JVM access. It will not run on serverless compute or Spark Connect clusters, because those environments block direct access to sparkContext and the low-level RDD API
```
